```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

## Functional method

```python
plt.plot(x,y,'r-')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('title')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('title')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel('X_label')
plt.ylabel('Y_label')
plt.title('title')

plt.tight_layout()
plt.show()
```

## OO

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.9,0.1,0.8,0.8])
axes1.plot(x,y,'r-')
axes1.set_xlabel('set xlabel')
axes1.set_ylabel('set ylabel')
axes1.set_title('title')



axes2.plot(y,x,'b-.')
axes2.set(xlabel='xlabel',ylabel='ylabel',title='set title')

plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].plot(x,y,'r')
ax[0,0].set_title('title')
ax[1,0].plot(y,x,'b')
ax[0,1].plot(x,y,'g')
ax[1,1].plot(y,x,'y')

plt.tight_layout()

plt.show()
```

## figsize,dpi,legend

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

for current_ax in ax:
    
    current_ax[1].plot(x,y)
    current_ax[0].plot(y,x)
    

plt.tight_layout()
    
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(5,4))

for current_ax in ax:
    
    current_ax[0].plot(x,x**2,label='squared')
    current_ax[1].plot(x,x**3,label='cubed')
    
plt.tight_layout()
plt.legend(loc=0)
plt.show()
```

```python
fig = plt.figure(figsize=(5,4))

axes = fig.add_axes([0,0,0.8,0.8])

axes.plot(x,x**2,label='sqared')
axes.plot(x,x**3,label= 'cubed')

plt.tight_layout()
plt.legend(loc=0)
plt.show()
```

## color, linewidth(lw),linestyle(ls),alpha,marker(0,+,*),markersize()

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

kwargs = dict()

axes.plot(x,y,color='green',linewidth=3,linestyle='-',alpha=0.9,marker='o',markersize=8,
         markerfacecolor='yellow',markeredgewidth=2,markeredgecolor='red')

plt.tight_layout()
plt.show()
```

## xlim,ylim

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlim([0,1])
axes.set_ylim([0,2])
plt.tight_layout()
plt.show()
```
